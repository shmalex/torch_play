```python
import numpy as np
from xml.dom import minidom
```

```python
# parse an xml file by name
mydoc = minidom.parse('./input/Afternoon_Hike.gpx')

data = []

for el in mydoc.getElementsByTagName('trkpt'):
    lat = float(el.attributes['lat'].value)
    lon = float(el.attributes['lon'].value)
    ele = float(el.childNodes[1].childNodes[0].data)
    data.append((lat, lon, ele))

data =np.array(data)
```

```python
min_lat, max_lat = np.min(data[:,0]), np.max(data[:,0])
min_lon, max_lon = np.min(data[:,1]), np.max(data[:,1])
```

```python
min_lat, max_lat
```

```python
# parse an xml file by name
mydoc = minidom.parse('./input/map.osm')

osm = []

for el in mydoc.getElementsByTagName('node'):
    lat = float(el.attributes['lat'].value)
    lon = float(el.attributes['lon'].value)
    if lat < min_lat or lat> max_lat: continue
    if lon < min_lon or lon> max_lon: continue
    osm.append((lat, lon, 170))

osm =np.array(osm)
print(len(osm))
```

```python
def trace(x,y,z,size, colors):
    #x, y, z = np.random.multivariate_normal(np.array([0,0,0]), np.eye(3), count).transpose()
    trace1 = go.Scatter3d(
        x=x,
        y=y,
        z=z,
        mode='markers',
        marker=dict(
            size=size,
            line=dict(
                color=colors, #'rgba(217, 217, 217, 0.14)',
                width=0.5
            ),
            opacity=0.8
        )
    )
    return trace1
```

```python
def show(datas, xrange, yrange):
    data=[trace(x,y,z,size, colors)for x,y,z,size, colors in datas]
    layout = go.Layout(
        margin=dict(
            l=0,
            r=0,
            b=0,
            t=0
        ),
        yaxis=dict(
            range=yrange
        ),
        xaxis=dict(
            range=xrange
        )
    )
    fig = go.Figure(data=data, layout=layout)
    iplot(fig, filename='simple-3d-scatter')
```

```python
show([(data[:,0]*(-1), data[:,1],data[:,2], 1, 'red'),(osm[:,0]*(-1), osm[:,1],osm[:,2], 1, 'blue')],
    (min_lat, max_lat), (min_lon, max_lon))
```
