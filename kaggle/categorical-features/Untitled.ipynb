```python
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
FILE='train'
```

```python
df = pd.read_csv(f'./input/{FILE}.csv')
```

```python
df.ord_2
```

```python
af = pd.read_csv('./input/animals.csv', sep='\t')
af = af.set_index('animal')
af_columns = af.columns
af = af.to_dict('index')
```

```python
af = defaultdict(str, af)
af[np.nan] = {'water': 'n', 'home': 'n', 'fure': 'n', 'mass': 7.0}
af
```

```python
af_columns
```

```python
df[['animal_' + afc for afc in af_columns]]=df.apply(lambda x: pd.Series(af[x['nom_2']]), axis=1)[af_columns]
```

```python
df.head()
```

```python
cf = pd.read_csv('input/countries.csv', sep='\t')
country_missing = cf.mean().to_dict()
cf = cf.set_index('country')
cf_columns = cf.columns
cf = cf.to_dict('index')
cf
```

```python
cf_columns
```

```python
country_missing
```

```python
cf = defaultdict(str, cf)
cf[np.nan] = country_missing
cf
```

```python
df[['country_'+cfc for cfc in cf_columns]]=df.apply(lambda x: pd.Series(cf[x['nom_3']]), axis=1)[cf_columns]
```

```python
df
```

```python
hf = pd.read_csv('input/heat.csv', sep='\t')
heat_missing = hf.mean().to_dict()
hf = hf.set_index('heat')
hf_columns = hf.columns
hf = hf.to_dict('index')
hf,heat_missing
```

```python
hf = defaultdict(str, hf)
hf[np.nan] = heat_missing
hf
```

```python
df[['heat_'+hfc for hfc in hf_columns]]=df.apply(lambda x: pd.Series(hf[x['ord_2']]), axis=1)[hf_columns]
```

```python
df
```

```python
inf = pd.read_csv('input/instruments.csv', sep='\t')
inst_missing = inf.mean().to_dict()
inst_missing = {'mass':2,'strings': 'n',
   'electric': 'n',
   'blow': 'y',
   'keys': 'n'}
inf = inf.set_index('instrument')
inf_columns = inf.columns
inf = inf.to_dict('index')
inf,inst_missing
```

```python
inf = defaultdict(str, inf)
inf[np.nan] = inst_missing
inf
```

```python
df[['inst_'+inc for inc in inf_columns]]=df.apply(lambda x: pd.Series(inf[x['nom_4']]), axis=1)[inf_columns]
```

```python
df
```

```python
rf = pd.read_csv('input/rank.csv', sep='\t')
rank_missing = rf.mean().to_dict()
rf = rf.set_index('title')
rf_columns = rf.columns
rf = rf.to_dict('index')
rf,rank_missing
```

```python
rf = defaultdict(str, rf)
rf[np.nan] = rank_missing
rf
```

```python
df[['rank_'+rfc for rfc in rf_columns]]=df.apply(lambda x: pd.Series(rf[x['ord_1']]), axis=1)[rf_columns]
```

```python
df
```

```python
df.to_csv(f'./input/{FILE}_expanded.csv', sep='\t', index=False)
```
